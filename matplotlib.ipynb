```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'apples', 'banans', 'grapes', 'mangoes'
print(type(labels))
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.1, 0.1)  # only "explode" the 2nd slice (i.e. 'bananas')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
         startangle=90)
 # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.axis('equal') 
plt.show()
```

# subplot
    Subplot function is used to define the plot position
    SYNTAX matplotlib.pyplot.subplot(row,column,position)
    Total Positions are given by row*column

```python
import numpy as np
import math
x=np.linspace(0,6*math.pi)
plt.subplot(2,2,1)  
plt.plot(x,np.sin(x),'ro')
plt.subplot(222)
plt.plot(x,np.sin(x),'g-')
plt.subplot(223)
plt.plot(x,np.cos(x),'ro')
plt.subplot(224)
plt.plot(x,np.cos(x),'g-')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
label = ['Monday','Tuesday','Wednseday','Thursday','Friday','Saturday','Sunday']
sales = [0,212,222,200,350,376,399]


index = np.arange(len(label))
plt.bar(index, sales)
plt.xlabel('Day', fontsize=20)
plt.ylabel('No of Units sold', fontsize=20)
plt.xticks(index, label, fontsize=15, rotation=20)
plt.title('Sales Data for each day of a week')

plt.show()

#HW add values with BARs
```

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped

A box and whisker plotalso called a box plotdisplays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum.


```python
import matplotlib.pyplot as plt


data=[25,25, 28,28,28, 29, 29, 30, 34, 35, 35, 37, 38]


plt.title('Basic Plot')
plt.boxplot(data,patch_artist=True)
plt.show()
```
