 ## pandas.Series()
 It is a one-dimensional labeled array capable of holding data of any type . It requires data in the form of list and index as another list

```python
import pandas as pd
a=pd.Series([1,'2',3],index=['a','b','c'])
print(a)
```

    
## pandas.DataFrame()
    This function will separate data into rows and columns.
    If there are no name specified for the row and column then it will by default label row and column as 0,1,2...
   ##### SYNTAX- pandas.DataFrame(list/data_name)

```python
import pandas as pd              
a=[[0,1,2,3],[4,5,6,7]]
s=pd.DataFrame(a)
print(s)
s.to_excel("demo.xlsx")
```

```python
import pandas as pd               
total=[]
#import pandas library
a=[['John',20,30],['William',60,70]]                                        #create a nested list this has two rows and four columns
s=pd.DataFrame(a,index=None,columns=['Student name','sub1','sub2'])#this will give name to the rows and columns
for i in range(0,len(s)):
    for j in range(1,len(s)):
        total.append(s.iloc[i,j]+s.iloc[i,j+1])
        print(total)
s['total']=total                 #adding new column

print(s)
 #this will give name to the rows and columns
s.to_excel('student.xlsx',index=None)
#this will save the code into the given location with excel file format. 
#index=none will remove the index from the saved excel file
```

Pandas treat None and NaN as essentially interchangeable for indicating missing or null values. To facilitate this convention, there are several useful functions for detecting, removing, and replacing null values in Pandas DataFrame :



isnull()
notnull()
dropna()
fillna()
replace()

```python
# importing pandas package  
import pandas as pd  
    
# making data frame from csv file  
data = pd.read_csv("employees.csv")  
    
# creating bool series True for NaN values  
bool_series = pd.isnull(data["Gender"])  
    
# filtering data  
# displaying data only with Gender = NaN  
data[bool_series]
#data.dropna()
data
```

```python
#Now we are going to fill all the null values in Gender column with â€œNo Genderâ€

# importing pandas package  
import pandas as pd  
    
# making data frame from csv file  
data = pd.read_csv("employees.csv")  
  
# filling a null values using fillna()  
data["Gender"].fillna("No Gender", inplace = True)  
  
data 
```

```python
#Dropping Rows with at least 1 null value in CSV file


# importing pandas module  
import pandas as pd  
    
# making data frame from csv file  
data = pd.read_csv("employees.csv")  
    
# making new data frame with dropped NA values  
new_data = data.dropna(axis = 0, how ='any')  
    
new_data 
```

```python
#Now we are going to replace the all Nan value in the data frame with -99 value.

# importing pandas package  
import pandas as pd  
    
# making data frame from csv file  
data = pd.read_csv("employees.csv")  
    
# will replace  Nan value in dataframe with value -99   
data.replace(to_replace = np.nan, value = -99) 
```
